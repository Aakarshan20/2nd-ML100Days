[範例重點]

    pandas.cut 的等寬劃分效果 (In[3], Out[4])
    pandas.qcut 的等頻劃分效果 (In[5], Out[6])



```python
# 載入套件
import os
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 初始設定 Ages 的資料
ages = pd.DataFrame({"age": [18,22,25,27,7,21,23,37,30,61,45,41,9,18,80,100]})
```

```python
ages
```

等寬劃分

```python
# 新增欄位 "equal_freq_age", 對年齡做等頻劃分
#qcut(x, q, labels=None, retbins=False, precision=3, duplicates='raise')
#q : integer or array of quantiles
#Number of quantiles. 10 for deciles, 4 for quartiles, etc. Alternately array of quantiles, e.g. [0, .25, .5, .75, 1.] for quartiles

ages["equal_freq_age"] = pd.qcut(ages["age"], 4)
```
